```python
## --- Get gradient intensity of all image covers. 
os.chdir(path+"/data/raw/covers")
image_files = [f for f in glob.glob('*.jpg')]
duplicates, ds_dict = get_gradient_intensity(image_files)
```

```python
# Build a distances matrix
hamm_matrix = [hamming_distance(ds_dict[image_files[col]], ds_dict[image_files[row]]) for col in range(len(image_files)) for row in range(len(image_files))]
hamm_matrix = np.array(hamm_matrix)
hamm_matrix.shape = (len(image_files), len(image_files))
hamm_df = pd.DataFrame(hamm_matrix)
hamm_df.head()
```

---

```python
## -- Variables
book_num = 2
top_num = 5
selected_rating = 3
# What are the top 5 similar covers for book 2
simil = hamm_df.iloc[:,book_num].sort_values(ascending=True).head(top_num+1).index.values[1:]
original = image_files[book_num]
book_id = data.loc[data.cover==image_files[book_num]].book_id.values[0]
data.loc[data.cover==image_files[book_num]]
```

```python
# create n subplots (hrztl stacked)
plt.imshow(imread(original))
plt.axis('off')
plt.title('Book Id: '+str(book_id))
plt.show()
f, axarr = plt.subplots(1,top_num,figsize=(20,5))
for idx,img in enumerate(simil):
#     print(idx,img)
    axarr[idx].imshow(imread(image_files[img]))
    axarr[idx].axis('off')
    simil_title = str(data.loc[data.cover==image_files[img]].book_id.values[0])\
                  + ": " + str(round(hamm_df.loc[simil[idx],book_num],5))
    axarr[idx].set_title(simil_title)

plt.show()
```

# A bit of network theory

```python
import networkx as nx
import re
from nltk.corpus import stopwords
import itertools

stop_words = stopwords.words('english')
thresh = 0.1 # Hamming distance threshold

# Build a list of lists such that: 
#     [ "book1, book2, hammingDistance>thresh",..,"book n-1, book n, hammingDistance>thresh"]
```

```python
## ----- PLOT Hamming Threshold vs Network Average Degree

a_D = []
for thresh in np.linspace(0.1, 1, num=30):
    # Keep those books linked with other books whose Hamming distance is above a certain threshold.
    nwk = [ str(k1)+' '+str(k2)+' 1' for k1, k2 in itertools.combinations(range(len(image_files)), 2) if hamming_distance(ds_dict[image_files[k1]], ds_dict[image_files[k2]])>thresh]
    # Parse into an ALL-Connected networkx
    G = nx.parse_edgelist(nwk, nodetype = int, data=(('weight',float),))
    # Get average degree
    avg_degree = [node[1] for node in list(G.degree())]
    norm_degree = np.mean(avg_degree) / (len(image_files)-1)
    a_D.append(norm_degree)
    
```

```python
plt.plot(np.linspace(0.1, 1, num=30), a_D)
plt.xlabel('Hamming threshold')
plt.ylabel("NTWK average degree %")
plt.grid()
plt.show()
```

```python
from IPython.core.magic import register_line_cell_magic

# Configuration parameters
max_show_lines = 50         # Limit on the number of lines to show with %sql_show and %sql_display
detailed_explain = True     # Set to False if you want to see only the physical plan when running explain


@register_line_cell_magic
def sql(line, cell=None):
    "Return a Spark DataFrame for lazy evaluation of the sql. Use: %sql or %%sql"
    val = cell if cell is not None else line 
    return sqlContext.sql(val)

@register_line_cell_magic
def sql_show(line, cell=None):
    "Execute sql and show the first max_show_lines lines. Use: %sql_show or %%sql_show"
    val = cell if cell is not None else line 
    return sqlContext.sql(val).show(max_show_lines) 

@register_line_cell_magic
def sql_display(line, cell=None):
    """Execute sql and convert results to Pandas DataFrame for pretty display or further processing.
    Use: %sql_display or %%sql_display"""
    val = cell if cell is not None else line 
    return sqlContext.sql(val).limit(max_show_lines).toPandas() 

@register_line_cell_magic
def sql_explain(line, cell=None):
    "Display the execution plan of the sql. Use: %sql_explain or %%sql_explain"
    val = cell if cell is not None else line 
    return sqlContext.sql(val).explain(detailed_explain)
```
